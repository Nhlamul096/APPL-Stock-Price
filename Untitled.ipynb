```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Download Apple stock data from yfinance
ticker = 'AAPL'
apple_stock = yf.Ticker(ticker)

# Fetch historical data (last 5 years)
df = apple_stock.history(period="5y")
```

```python
# Plot the closing prices over time
plt.figure(figsize=(10, 6))
plt.plot(df['Close'], label='Closing Price')
plt.title('Apple Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create feature and target
df['Target'] = df['Close'].shift(-1)  # Predict the next day's price
df = df.dropna()  # Remove rows with missing values

X = df[['Close']].values  # Features (close price)
y = df['Target'].values   # Target (next day price)
```

```python
from sklearn.model_selection import train_test_split

# Split the data: 80% for training, 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
from sklearn.linear_model import LinearRegression

# Initialize and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)

# Plot the actual vs predicted prices
plt.figure(figsize=(10, 6))
plt.plot(df.index[len(X_train):], y_test, label='Actual Prices')
plt.plot(df.index[len(X_train):], y_pred, label='Predicted Prices')
plt.title('Actual vs Predicted Prices')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Predict the next 30 days of prices
future_days = 30
last_price = X_test[-1].reshape(1, -1)  # Get the last closing price

future_prices = []
for _ in range(future_days):
    next_price = model.predict(last_price)[0]
    future_prices.append(next_price)
    last_price = np.array([[next_price]])

# Generate future dates for the next 30 days
future_dates = pd.date_range(df.index[-1], periods=future_days + 1)

# Plot future predictions
plt.figure(figsize=(10, 6))
plt.plot(future_dates[1:], future_prices, label='Predicted Future Prices')  # Skip the first date as it's the current date
plt.title('Apple Stock Price Prediction for Next 30 Days')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()

```
